# 1) 전처리

```python
import pandas as pd
import numpy as np
import re
df1 = pd.read_csv('/content/drive/MyDrive/데이터/빅데이터개론/lipstick_data.csv')
df1.head()
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/데이터/빅데이터개론/liptint_data.csv')
df2.head()
```

```python
df1['찜하기개수'] = df1['Text'].str.extract(r'(\d+)')
df1 = df1.astype({'찜하기개수':'int'})

df1['Price'] = df1['Price'].str.extract(r'(\d+)')
df1 = df1.astype({'Price':'int'})
```

```python
df2['찜하기개수'] = df2['Text'].str.extract(r'(\d+)')
df2 = df2.astype({'찜하기개수':'int'})

df2['Price'] = df2['Price'].str.extract(r'(\d+)')
df2 = df2.astype({'Price':'int'})
```

```python
red = []
for i in range(len(df1['Detail'])):
  a = [word for word in df1['Detail'][i].split(':') if '레드' in word]
  if len(a) == 0:
    red.append(0)
  else:
    red.append(len(a[0].split(',')))

pink = []
for i in range(len(df1['Detail'])):
  a = [word for word in df1['Detail'][i].split(':') if '핑크' in word]
  if len(a) == 0:
    pink.append(0)
  else:
    pink.append(len(a[0].split(',')))
```

```python
df1['red'] = red
df1['pink'] = pink

color = []
for i in range(len(df1['Detail'])):
  if df1['red'][i] == df1['pink'][i]:
    color.append(df1['red'][i])
  else:
    color.append(df1['red'][i] + df1['pink'][i])
```

```python
df1['color'] = color
```

```python
red = []
for i in range(len(df2['Detail'])):
  a = [word for word in df2['Detail'][i].split(':') if '레드' in word]
  if len(a) == 0:
    red.append(0)
  else:
    red.append(len(a[0].split(',')))

pink = []
for i in range(len(df2['Detail'])):
  a = [word for word in df2['Detail'][i].split(':') if '핑크' in word]
  if len(a) == 0:
    pink.append(0)
  else:
    pink.append(len(a[0].split(',')))
```

```python
df2['red'] = red
df2['pink'] = pink

color = []
for i in range(len(df2['Detail'])):
  if df2['red'][i] == df2['pink'][i]:
    color.append(df2['red'][i])
  else:
    color.append(df2['red'][i] + df2['pink'][i])
```

```python
df2['color'] = color
```

```python
def inword(x):
  if '촉촉' in x:
    x = 1
  else:
    x = 0
  return x
```

```python
df1['수분감'] = df1['Detail'].apply(inword)
df2['수분감'] = df2['Detail'].apply(inword)
```

```python
df1.drop(['Detail', 'Text', 'red', 'pink'], inplace = True, axis = 1)
df2.drop(['Detail', 'Text', 'red', 'pink'], inplace = True, axis = 1)
```

```python
df1['립스틱여부'] = 1
df2['립스틱여부'] = 0
```

```python
df1.columns = ['상품명', '가격', '찜하기개수', '색', '수분감', '립스틱여부']
df2.columns = ['상품명', '가격', '찜하기개수', '색', '수분감', '립스틱여부']
```
